```python
step: int = 3
def compute_indices_to_keep(b, step) -> list[int]:
    start = step-1
    indices_to_keep = [start]

    s = 1
    while indices_to_keep[-1] < len(b):
        indices_to_keep.append(start + step*s)
        s+=1
    indices_to_keep.pop()

    indices_to_keep

compute_indices_to_keep(step)

```

```python
def compute_indices_to_remove(b, indices_to_keep):
    return list(set(range(len(b))).difference(indices_to_keep))

indices_to_remove = compute_indices_to_remove()
indices_to_remove
```

```python
np.delete(b, indices_to_remove)
```

```python
def trim(data: np.ndarray, n: int = 3) -> np.ndarray:
    kept_indices: list[int] = compute_indices_to_keep(data, n)

    removed_indices: list[int] = compute_indices_to_remove(data, kept_indices)

    if len(data.shape) == 2:
        return np.delete(data, removed_indices, 1)

    else:
        return np.delete(data, removed_indices)


```
